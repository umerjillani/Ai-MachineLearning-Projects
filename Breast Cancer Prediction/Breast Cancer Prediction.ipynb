```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('breast-cancer.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.describe()
```

```python
X = df.drop(['id', 'diagnosis'], axis=1) 
y = df['diagnosis'].map({'M':1, 'B':0}) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Random Forest

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model_predictions = model.predict(X_test)
```

```python
model_predictions
```

```python
print(f"Random Forest Accuracy: {np.round(accuracy_score(y_test, model_predictions), 2) * 100} %\n")
print(classification_report(y_test, model_predictions)) 
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model2 = LogisticRegression()
```

```python
model2.fit(X_train, y_train)
```

```python
model2_predictions = model2.predict(X_test)
```

```python
print(f"Logistic Regression Accuracy: {np.round(accuracy_score(y_test, model2_predictions), 3)*100} %\n") 
print(classification_report(y_test, model2_predictions))
```
